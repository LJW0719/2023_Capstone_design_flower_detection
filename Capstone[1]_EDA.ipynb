```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from torchvision import datasets
```

## Show Image

```python
# set sunflower path
sunflower_paths=[]
for dirname, _, filenames in os.walk('/content/drive/MyDrive/flowers/sunflower'):
    for filename in filenames:
        sunflower_paths += [(os.path.join(dirname, filename))]
```

```python
# set flower path
flower_paths=[]
for dirname, _, filenames in os.walk('/content/drive/MyDrive/flowers'):
    for filename in filenames:
        flower_paths += [(os.path.join(dirname, filename))]
```

```python
# show sunflower image
f, ax = plt.subplots(3, 3, figsize=(10,10))
for i in range(9):
    c=i%3
    r=i//3
    path=sunflower_paths[i]
    img=cv2.imread(path)
    ax[r][c].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[r][c].set_xticks([])
    ax[r][c].set_yticks([])
plt.show()
```

```python
# show flower image
f, ax = plt.subplots(3, 3, figsize=(10,10))
for i in range(9):
    c=i%3
    r=i//3
    path=flower_paths[i]
    img=cv2.imread(path)
    ax[r][c].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[r][c].set_xticks([])
    ax[r][c].set_yticks([])
plt.show()
```
